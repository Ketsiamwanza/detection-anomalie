```python
from scapy.all import sniff
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns
# Compteur
from collections import Counter

# Analyse des performances
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay, PrecisionRecallDisplay
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, classification_report, RocCurveDisplay, roc_curve, auc
from sklearn import metrics

# Data spliting
from sklearn.model_selection import train_test_split

# Mise à l'échelle des données (Normalisation des données)
from sklearn.preprocessing import StandardScaler
```

```python
# Liste pour stocker les paquets capturés
captured_packets = []


def packet_callback(packet):
    # Afficher le paquet capturé
    packet.show()

    # Extraire les informations pertinentes du paquet
    packet_info = {
        'timestamp': packet.time,
        'source': packet[0][1].src,
        'destination': packet[0][1].dst,
        'protocol': packet[0][1].proto,
        'length': len(packet)
    }
    captured_packets.append(packet_info)

# Capture des paquets
print("Capture de paquets en cours... (Appuyez sur Ctrl+C pour arrêter)")
try:
    sniff(prn=packet_callback, store=0)  # store=0 évite de stocker les paquets en mémoire
except KeyboardInterrupt:
    print("Capture interrompue.")
```

```python
# Convertir la liste en DataFrame après la capture
df = pd.DataFrame(captured_packets)

# Enregistrer les données dans un fichier CSV
df.to_csv('captured_packets.csv', index=False)
print("Données capturées enregistrées dans 'captured_packets.csv'")
```

```python
# Charger les données à partir d'un fichier CSV
data = pd.read_csv("captured_packets.csv")
```

```python
# Nettoyer les données
data = data.dropna()  # Supprimer les valeurs manquantes
```

## Construction du modèle

```python
# Normalisation des caractéristiques
data['normalized_length'] = (data['length'] - data['length'].mean()) / data['length'].std()
```

```python
# Créer et entraîner le modèle
model = IsolationForest(contamination=0.1)  # Ajustez le taux de contamination
model.fit(data[['normalized_length']])
```

```python
# Prédire les anomalies
data['anomaly'] = model.predict(data[['normalized_length']])
```

```python
# -1 indique une anomalie, 1 indique un comportement normal
anomalies = data
#print("Anomalies détectées :")
#print(anomalies)
```

```python
X_train = pd.DataFrame(X_train)
```

```python
X_test = pd.DataFrame(X_test)
```

```python
# Créer et entraîner le modèle
model = IsolationForest(contamination=0.1)  # Ajustez le taux de contamination
model.fit(X_train)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# Calculer la courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
```

```python
# Tracer la courbe ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label='Courbe ROC (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Diagonale
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Courbe ROC pour le modèle d\'Isolation Forest')
plt.legend(loc='lower right')
plt.grid()
plt.savefig("roc.png",dpi=200)
plt.show()
```

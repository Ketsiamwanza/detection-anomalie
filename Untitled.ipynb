```python
from scapy.all import sniff
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
# Liste pour stocker les paquets capturÃ©s
captured_packets = []


def packet_callback(packet):
    # Afficher le paquet capturÃ©
    packet.show()

    # Extraire les informations pertinentes du paquet
    packet_info = {
        'timestamp': packet.time,
        'source': packet[0][1].src,
        'destination': packet[0][1].dst,
        'protocol': packet[0][1].proto,
        'length': len(packet)
    }
    captured_packets.append(packet_info)

# Capture des paquets
print("Capture de paquets en cours... (Appuyez sur Ctrl+C pour arrÃªter)")
try:
    sniff(prn=packet_callback, store=0)  # store=0 Ã©vite de stocker les paquets en mÃ©moire
except KeyboardInterrupt:
    print("Capture interrompue.")
```

```python
# Convertir la liste en DataFrame aprÃ¨s la capture
df = pd.DataFrame(captured_packets)

# Enregistrer les donnÃ©es dans un fichier CSV
df.to_csv('captured_packets.csv', index=False)
print("DonnÃ©es capturÃ©es enregistrÃ©es dans 'captured_packets.csv'")
```

```python
# Charger les donnÃ©es Ã  partir d'un fichier CSV
data = pd.read_csv("captured_packets.csv")
```

```python
# Nettoyer les donnÃ©es
data = data.dropna()  # Supprimer les valeurs manquantes
```

```python
plt.figure(figsize=(8,5))
pd.crosstab(index=data["protocol"],columns=data["anomaly"].sort_values(ascending=False)).plot(kind='barh',stacked=True,figsize=(8,5))
```

## Construction du modÃ¨le

```python
# Normalisation des caractÃ©ristiques
data['normalized_length'] = (data['length'] - data['length'].mean()) / data['length'].std()
```

```python
# CrÃ©er et entraÃ®ner le modÃ¨le
model = IsolationForest(contamination=0.1)  # Ajustez le taux de contamination
model.fit(data[['normalized_length']])
```

```python
# PrÃ©dire les anomalies
data['anomaly'] = model.predict(data[['normalized_length']])
```

```python
# -1 indique une anomalie, 1 indique un comportement normal
anomalies = data[data['anomaly'] == -1]
print("Anomalies dÃ©tectÃ©es :")
print(anomalies)
```

```python
anomalies_df = pd.DataFrame(anomalies)
```

```python
# Visualiser les anomalies
plt.figure(figsize=(10, 6))
plt.scatter(data['length'], data['anomaly'], c=data['anomaly'], cmap='coolwarm', alpha=0.5)
plt.title('DÃ©tection des anomalies de trafic rÃ©seau')
plt.xlabel('Longueur du paquet')
plt.ylabel('Anomalie')
plt.axhline(0, color='red', linestyle='--')  # Ligne de seuil
plt.grid()
plt.savefig("resultat-anomalie.png",dpi=200)
plt.show()
```
